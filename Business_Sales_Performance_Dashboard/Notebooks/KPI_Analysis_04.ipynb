# Business Sales Performance Analytics

## KPI Analysis

### Objective

Calculate key business performance indicators (KPIs) to support executive decision-making and dashboard development.

## 1. Load Dataset

```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns",None)

df=pd.read_csv("../outputs/cleaned_superstore.csv", parse_dates=["Order Date", "Ship Date"] )

df.head()
              
```

## 2. Data Verification

```python
print("Dataset shape:", df.shape)
```

## 3. Total Sales Revenue

```python
total_sales = df["Sales"].sum()
print("Total Sales Revenue: ${:,.2f}".format(total_sales))
```

## 4. Total Profit

```python
total_profit = df["Profit"].sum()
print("Total Profit: ${:,.2f}".format(total_profit))
```

## 5. Total Orders

```python
total_orders = df["Order ID"].nunique()
print("Total Orders:",total_orders)
```

## 6. Average Order Value (AOV)

```python
average_order_value = ( 
    total_sales / total_orders
)
print("Average Order Value: {:,.2f}".format(average_order_value))
```

## 7. Overall Profit Margin

```python
profit_margin = ( 
    total_profit / total_sales
) * 100
print("Profit Margin: {:,.2f}%".format(profit_margin))
```

## 8. Best Performing Category

```python
best_category = ( df.groupby("Category")["Profit"].sum().idxmax() )
print("Best Category:",best_category)
```

## 9. Best Performing Region

```python
best_region = ( df.groupby("Region")["Profit"].sum().idxmax() )
print("Best Region:",best_region)
```

## 10.Best Performing Product

```python
best_product = ( df.groupby("Product Name")["Profit"].sum().idxmax() )
print("Best Product:",best_product)
```

## 11. KPI Summary

The following KPIs provide a high-level overview of business performance and will be used in the Executive Dashboard.

```python
kpi_summary = pd.DataFrame({
    "KPI": [
        "Total Sales Revenue",
        "Total Profit",
        "Total Orders",
        "Average Order Value",
        "Profit Margin (%)",
        "Best Category",
        "Best Region",
        "Best Product"
        ],
    "Value" : [
        round(total_sales, 2),
        round(total_profit, 2),
        total_orders,
        round(average_order_value, 2),
        round(profit_margin, 2),
        best_category,
        best_region,
        best_product
    ]
})

kpi_summary
```

## 12. Export KPI Summary

```python
kpi_summary.to_csv("../outputs/kpi_summary.csv",index=False)
print("KPI Summary exported sucessfully")
```
